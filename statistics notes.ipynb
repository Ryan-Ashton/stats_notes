# Sample Size Determination

```python
# Two Proportion Z test

# Calculate a difference between p1 and p2 
# The sample required to determine differences between p1 and p2 depend on the absolute level of p1

# https://towardsdatascience.com/how-to-use-python-to-figure-out-sample-sizes-for-your-study-871f0a76a19c

import math

def z_calc(p1, p2, n1, n2):
    p_star = (p1*n1 + p2*n2) / (n1 + n2)
    return (p2 - p1) / math.sqrt(p_star*(1 - p_star)*((1.0 / n1) + (1.0 / n2)))

def sample_required(p1, p_diff, alpha):
    if p_diff <= 0:
        raise ValueError("p_diff must be > 0")
    n = 1
    while True:
        z = z_calc(p1, p1+p_diff, n1=n, n2=n)
        p = 1 - stats.norm.cdf(z)
        if p < alpha:
            break
        n += 1
    return n

sample_required(p1=.9, p_diff=.05, alpha=.05)


# Cochran’s formulae

# https://towardsdatascience.com/how-to-sample-data-with-code-327359dce10b

def sample(N, cl, e, p):
    # first we get the z-score
    z = stats.norm.ppf(1 - (1 - cl) / 2)
    # then the n_0 value
    n_0 = z**2 * p * (1 - p) / e**2
    # and finally we calculate n
    n = n_0 / (1 + (n_0 - 1) / N)
    # we also need to round up to the nearest integer
    n = math.ceil(n)
    # finally we return our sample size
    return n


# population size: N = 100000
# confidence level: 0.95
# margin of error: 0.05
# target proportion: 0.5

sample(100000, 0.95, 0.05, 0.5)
```

```python
# homogeneity of variance or homoscedasticity

# If the p-value is >0.05, the obtained differences in sample variances are unlikely to have occurred based on 
# random sampling from a population with equal variances. 

test_stat_var_one, p_value_var_one = stats.levene(quant_continuous_norm_one,quant_continuous_norm_two)
test_stat_var_two, p_value_var_two = stats.levene(quant_continuous_norm_two,quant_continuous_norm_three)
test_stat_var_three, p_value_var_three = stats.levene(quant_continuous_norm_one,quant_continuous_norm_three)

test_stat_var_four, p_value_var_four = stats.levene(quant_continuous_not_norm_one,quant_continuous_not_norm_two)
test_stat_var_five, p_value_var_five = stats.levene(quant_continuous_not_norm_two,quant_continuous_not_norm_three)
test_stat_var_six, p_value_var_six = stats.levene(quant_continuous_not_norm_one,quant_continuous_not_norm_three)

print("Quant Dataset 1: " + str(p_value_var_one))
print("Quant Dataset 2: " + str(p_value_var_two))
print("Quant Dataset 3: " + str(p_value_var_three))
print("Quant Dataset 4: " + str(p_value_var_four))
print("Quant Dataset 5: " + str(p_value_var_five))
print("Quant Dataset 6: " + str(p_value_var_six))
```

```python
# T-Test One Sample

t_stat, pval = stats.ttest_1samp(quant_continuous_norm_one,
                  popmean=49,
                  axis=0,
                  nan_policy='propagate',
                  alternative='greater')

print(t_stat, pval)
```

```python
# Unpaired T Test

statistic, pval = stats.ttest_ind(a=quant_continuous_norm_one,
                      b=quant_continuous_norm_two,
                      axis=0,
                      equal_var=True,
                      nan_policy='propagate',
                      permutations=None,
                      random_state=None,
                      alternative='less',
                      trim=0)

print(statistic, pval)
```

```python
# Paired T Test

statistic, pval = stats.ttest_rel(a=quant_continuous_norm_one,
                b=quant_continuous_norm_two,
                axis=0,
                nan_policy='propagate',
                alternative='less')

print(statistic, pval)
```

```python
# Repeated Measures ANOVA

from statsmodels.stats.anova import AnovaRM
```

```python
# One-Way ANOVA

statistic, pval = stats.f_oneway(quant_continuous_norm_one,
                quant_continuous_norm_two,
                quant_continuous_norm_three, axis=0)

print(statistic, pval)
```

```python
# Wilcoxon

statistic, pval = stats.wilcoxon(x=quant_continuous_not_norm_one,
                                               y=quant_continuous_not_norm_two,
                                               zero_method='wilcox',
                                               correction=False,
                                               alternative='two-sided',
                                               method='auto',
                                               axis=0,
                                               nan_policy='propagate',
                                               keepdims=False)

print(statistic, pval)
```

```python
# Friedman

statistic, pval = stats.friedmanchisquare(quant_continuous_not_norm_one,
                       quant_continuous_not_norm_two,
                       quant_continuous_not_norm_three)

print(statistic, pval)
```

```python
# Mann-Whitney U Test

statistic, pval = stats.mannwhitneyu(x=quant_continuous_not_norm_one,
                                     y=quant_continuous_not_norm_two,
                                     use_continuity=True,
                                     alternative='two-sided',
                                     axis=0,
                                     method='auto',
                                     nan_policy='propagate',
                                     keepdims=False)

print(statistic, pval)
```

```python
# Kruskal-Wallis H Test

statistic, pval =stats.kruskal(quant_continuous_not_norm_one, quant_continuous_not_norm_two, quant_continuous_not_norm_three,
              nan_policy='propagate',
              axis=0,
              keepdims=False)

print(statistic, pval)
```
